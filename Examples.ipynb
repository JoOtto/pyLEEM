## Examples for the pyLEEM package

The following notebook shows some examples on the usage of the pyLEEM package. The package right is predominantly useful for loading nlp files with python which are produced by the SPECS FE-LEEM P90.

### 1. Basic File Loading

After installing the library as explained in the readme.md on the GitHub page, the following code snippet shows the main workflow on loading files. As an example, we will load the ESCHER file provided in the test directory.

```python
import xarray
measurement = xarray.open_dataset("tests/data/20190223_185646_5.7um_349.0_test_ESCHER.nlp",engine='pyLEEM') 
```

### 2. Explore the Data

We can now display the content of the measurement and explore the meta data.

```python
measurement
```

The data structure displayed is called an xarray, which is basically numpy array with metadata. We can now access different metadata. Frame-specific metadata can be found in the data variables and accessed by the index or the time the frame was recorded. The `max_counts`, refering to the maximum measured pixel value in the frame, are given for the first frame as:

```python
measurement.max_counts.values[0]
```

Metadata which describes the full image is saved in the xarray attributes and can be accessed via the `attrs` property. We can access the gun value from L5 set for the different images as:

```python
measurement.attrs['GUN']
```

### 3. View the images

For pyLEEM we added napari in the requirements for easy analysis of the LEEM images. The follwing example shows the usage of napari together with the loaded xarray to display the image(s). **Napari will open in another window.**

```python
import napari
viewer = napari.view_image(measurement.intensity)
```
